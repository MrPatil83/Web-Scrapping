```python
import requests
import datetime
import pandas as pd

current_date = datetime.datetime.now()

date_time = current_date.strftime('%d-%b-%Y')


year=[2016,2017,2018]
for l in year:
    l=str(l)
    link = "https://buy.iciciprulife.com/buy/funds-nav-history.htm?fundCode=PGTH&startDate=01-Jan-"+l+"&endDate=30-Jun-"+l
    print(link)

    response = requests.get(link)

    A=[]
    B=[]
    if response.status_code == 200:
        result = response.json()
        print(result)
        for i in result:
            date= i["x"]
            A.append(date)
            nav_rate= i["y"]
            B.append(nav_rate)
            
    df=pd.DataFrame(B,A)
    print(df)

    filename="1-jan to 30-june"+l+".csv"
    print(filename)
    df.to_csv(filename)


for l in year:
    l=str(l)
    link = "https://buy.iciciprulife.com/buy/funds-nav-history.htm?fundCode=PGTH&startDate=01-Jul-"+l+"&endDate=31-Dec-"+l
    print(link)

    response = requests.get(link)

    A=[]
    B=[]
    if response.status_code == 200:
        result = response.json()
        print(result)
        for i in result:
            date= i["x"]
            A.append(date)
            nav_rate= i["y"]
            B.append(nav_rate)
            
    df=pd.DataFrame(B,A)
    print(df)

    filename="1-Jul To 31-Dec"+l+".csv"
    print(filename)
    df.to_csv(filename)
```
