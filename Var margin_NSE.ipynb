```python
import datetime
import requests
import os
import time

filepath = r'M:\Secured\ENS Funds\Mutual Fund\Data\Economy Tool Data\Raw\NSE\Var Margin'

class Vardownload:

    def __init__(self):
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36',
            'Referer': 'https://archives.nseindia.com/',
        }
        self.session = requests.Session()
        self.session.headers.update(self.headers)
        self.session.mount('https://', requests.adapters.HTTPAdapter(max_retries=5))

    def download_file(self, file_url, local_file_path):
        try:
            response = self.session.get(file_url)
            response.raise_for_status()

            with open(local_file_path, 'wb') as file:
                file.write(response.content)
            print(f"Done --- {local_file_path}")
        except (requests.exceptions.RequestException, IOError) as e:
            print(f"Error downloading {file_url}: {e}")

    def create_link(self):
        for i in range(5, 7):

            current_date = datetime.datetime.now()
            start_date = datetime.datetime(2007, 1, 1)

            delta = datetime.timedelta(days=1)
            while start_date <= current_date:
                if start_date.weekday() not in (5, 6):
                    date_str = start_date.strftime("%d%m%Y")
                    link = f'https://archives.nseindia.com/archives/nsccl/var/C_VAR1_{date_str}_{i}.DAT'

                    directory_name = f'Data_{date_str}'
                    download_path = os.path.join(filepath, directory_name)
                    os.makedirs(download_path, exist_ok=True)

                    file_name = f'C_VAR1_{date_str}_{i}.DAT'
                    download_path = os.path.join(download_path, file_name)

                    self.download_file(link, download_path)
                    time.sleep(1)
                start_date += delta

file = Vardownload()
file.create_link()
# datetime.datetime.now()
```
