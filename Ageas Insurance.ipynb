```python
import requests
from bs4 import BeautifulStoneSoup
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import requests
import os
download_path=r"C:\Users\Avinash\OneDrive\Desktop\Demo\Ageas"
chrome_option = Options()
chrome_option.add_experimental_option('prefs', {
                'download.default_directory': download_path,
                'download.prompt_for_download': False,
                'download.directory_upgrade': True,
                'safebrowsing.enabled': True
            })
 
driver = webdriver.Chrome(options=chrome_option)


url = "https://www.ageasfederal.com/public-disclosures"
driver.get(url)
driver.maximize_window()
#click on year box
driver.find_element(By.XPATH,"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[1]/div/div[1]/select").click()
# time.sleep(2)
#select year list number(start year column 2)
driver.find_element(By.XPATH,"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[1]/div/div[1]/select/option[6]").click()####change
time.sleep(2)
# #click month box
driver.find_element(By.XPATH,"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[1]/div/div[2]/select").click()
time.sleep(2)
# #select Quater Month(start month 2,3,4,5)
driver.find_element(By.XPATH,"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[1]/div/div[2]/select/option[5]").click()###change
time.sleep(2)
# #click On filter Button
driver.find_element(By.XPATH,"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[1]/div/div[3]/button").click()
time.sleep(2)
#click to download
for n in range(1,44):
        l=f"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[2]/ul/li[{n}]/div/a[1]"
        # l=f"/html/body/div/div[5]/div/div[3]/div/div[1]/div/div[2]/ul/li[{n}]/div/a[2]"
        pdf=driver.find_element(By.XPATH,l).get_attribute('href')
        file_name = pdf.split("/")[-1]

        # Download the file only if it doesn't exist
        local_file_path = os.path.join(download_path, file_name)
        if not os.path.exists(local_file_path):
            response = requests.get(pdf)
            if response.status_code == 200:
                with open(local_file_path, "wb") as file:
                    file.write(response.content)
                print(f"Download successful. File saved at: {local_file_path}")
            else:
                print(f"Failed to download. Status code: {response.status_code}")
        else:
            print(f"File already exists. Skipping download.")
            
            
            
            
            
            
# get_attribute('href')
# def download_file(file_url):
#         response = requests.get(file_url)
#         if response.status_code == 200:
#             excel_data = response.content
#             local_file_path = 'downloaded_file.pdf'
#             with open(local_file_path, 'wb') as file:
#               file.write(excel_data)            
```
